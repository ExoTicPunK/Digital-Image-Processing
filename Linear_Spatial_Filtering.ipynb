# Linear Spatial Filtering


# BOX FILTER

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import convolve
```

```python
def box_filter(image, kernel_size):
  kernel = np.ones((kernel_size, kernel_size)) / (kernel_size**2)
  filtered_image = convolve(image, kernel, mode='reflect')
  return filtered_image
```

```python
image_path = '/content/drive/MyDrive/Digital-Image-Processing/img5.jpg'
image=np.array(Image.open(image_path).convert('L'))
```

```python
box_filtered_image=box_filter(image,kernel_size=10)
```

```python
box_filtered_image2=box_filter(image,kernel_size=20)
```

```python
box_filtered_image3=box_filter(image,kernel_size=40)
```

```python
fig, axs = plt.subplots(1, 4, figsize=(12, 6))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(box_filtered_image, cmap='gray')
axs[1].set_title('Box Filtered Image 10X10')
axs[1].axis('off')

axs[2].imshow(box_filtered_image2, cmap='gray')
axs[2].set_title('Box Filtered Image 20X20')
axs[2].axis('off')

axs[3].imshow(box_filtered_image3, cmap='gray')
axs[3].set_title('Box Filtered Image 40X40')
axs[3].axis('off')

plt.show()
```

# Gaussian Filter

```python
from scipy.ndimage import gaussian_filter as g_f
```

```python
def gaussian_filter(image, sigma):
    filtered_image = g_f(image, sigma=sigma)
    return filtered_image
```

```python
gaussian_filtered_image = gaussian_filter(image, sigma=10)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(gaussian_filtered_image, cmap='gray')
axs[1].set_title('Gaussian Filtered Image')
axs[1].axis('off')

plt.show()
```

# ORDER STATSTICS FILTER (Non-Linear filter)

```python
import matplotlib.pyplot as plt
from scipy.ndimage import median_filter, maximum_filter, minimum_filter
import numpy as np
```

```python
if image.ndim == 3:
  image = np.mean(image, axis=2)
```

```python
median_filtered_image = median_filter(image, size=10)
max_filtered_image = maximum_filter(image, size=10)
min_filtered_image = minimum_filter(image, size=10)
```

```python
fig, axs = plt.subplots(1, 4, figsize=(18, 5), sharex=True, sharey=True)
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(median_filtered_image, cmap='gray')
axs[1].set_title('Median Filter')
axs[1].axis('off')

axs[2].imshow(max_filtered_image, cmap='gray')
axs[2].set_title('Maximum Filter')
axs[2].axis('off')

axs[3].imshow(min_filtered_image, cmap='gray')
axs[3].set_title('Minimum Filter')
axs[3].axis('off')

plt.tight_layout()
plt.show()
```

# Shading correction using lowpass filtering.

    image:- Hubble Telescope image of the Hickson Compact Group.
    (Original image courtesy of NASA.)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter as g_f
from skimage import io, img_as_float
from skimage.filters import threshold_otsu
```

```python
image_path = '/content/drive/MyDrive/Digital-Image-Processing/img6.png'
image=np.array(Image.open(image_path).convert('L'))
image = img_as_float(image)
```

```python
sigma = 5
filtered_image = g_f(image, sigma=sigma)
```

```python
thresh = threshold_otsu(filtered_image)
binary_image = filtered_image > thresh
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(filtered_image, cmap='gray')
axs[1].set_title('Lowpass Filtered Image')
axs[1].axis('off')

axs[2].imshow(binary_image, cmap='gray')
axs[2].set_title('Thresholded Image')
axs[2].axis('off')

plt.show()
```
